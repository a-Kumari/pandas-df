```python
import numpy as np 
import pandas as pd 
titanic = pd.read_csv('/home/development/Downloads/practice/titanic.csv')
titanic.head()
```

```python
titanic['Name']
```

```python
# Common Functions
# lower/upper/capitalize/title
titanic['Name'].str.lower()
titanic['Name'].str.upper()
# titanic['Name'].str.capitalize()
# titanic['Name'].str.title()
```

```python
# len
titanic['Name'][titanic['Name'].str.len()==82].values[0]
```

```python
# strip
titanic['Name'].str.strip()
```

```python
#split---> get
titanic['Lastname']=titanic['Name'].str.split(',').str.get(0)
titanic.head()
```

```python
titanic[['Title', 'Firstname']]=titanic['Name'].str.split(',').str.get(1).str.strip().str.split(' ', n=1, expand=True)
titanic.head()
```

```python
# replace()
titanic['Title'].value_counts()
titanic['Title']=titanic['Title'].str.replace('Ms.', 'Miss.')
titanic['Title']=titanic['Title'].str.replace('Mlle.', 'Miss.')
titanic['Title'].value_counts()
```

```python
# filtering
# startswith/endswith
titanic[titanic['Name'].str.startswith('A')]
# isdigit/isalpha...
titanic[titanic['Lastname'].str.isdigit()]
```
