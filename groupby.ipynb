```python
# looping on groups
# find the highest rated movie of each genre
for groups,gen in genre:
    df = gen[gen['IMDB_Rating'] == gen['IMDB_Rating'].max()]
    print(df)
```

```python
# split (apply) combine
#apply()
# find number of movies starting with A for each group
def start_a(group):
    return group['Series_Title'].str.startswith('A').sum()

genre.apply(start_a, include_groups=False)
```

```python
# find ranking of each movie in the group according to IMDB score

def movie_rank(group):
    group['Movie_Rank']= group['IMDB_Rating'].rank(ascending=False)
    return group

genre.apply(movie_rank, include_groups=False)
```

```python
# find normalized IMDB rating group wise
def normalized(group):
    group['Normal_Rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group

genre.apply(normalized, include_groups=False)
```

```python
ipl = pd.read_csv('/home/development/Downloads/practice/deliveries.csv')
ipl.head()
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of sixes
six = ipl[ipl['batsman_runs'] == 6]
six.groupby('batsman')['batsman'].value_counts().sort_values(ascending=False).head(1)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
last_over = ipl[ipl['over'] > 15]
over =last_over[(last_over['batsman_runs'] == 4 ) | (last_over['batsman_runs'] == 6)]
over.groupby('batsman')['batsman'].value_counts().sort_values(ascending=False).head(1).index[0]
```

```python
# find V Kohli's record against all teams
vk = ipl[ipl['batsman'] == 'V Kohli']
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
# Create a function that can return the highest score of any batsman
def highest_score(batsman):
    batsman = ipl[ipl['batsman'] == batsman]
    return batsman.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0].item()

highest_score('CH Gayle')
```

```python
# Questions of groupby
train = pd.read_csv('/home/development/Downloads/practice/train-train.csv')
train.head()
```

```python
# Q-1: Using groupby make groups using the "Pclass" column and 
# find out the average age and total number of missing values in the "Age" column for every group.

pclass = train.groupby('Pclass')
age_avg = pclass['Age'].mean()
age_avg
# age_nan = pclass['Age'].apply(lambda x: x.isnull().sum())
# age_nan
```

```python
# Q-2: Using groupby make groups using the "Pclass" column and fill every group's "Embarked" column's missing values with the mode value of that group. 
# After that, print every group's "Embarked" column's value counts in ascending order.
train.groupby('Pclass')['Embarked'].apply(lambda x: x.fillna(x.mode()[0])).value_counts(ascending=False)
```

```python
# Q-3: Make groups based on "Embarked" column. And for each of this embarked group, make another group based on "Pclass" and 
# find out the average fare (round off up to 2 decimal places) for each "Pclass" for each group of "Embarked".
avg_fare = train.groupby(['Embarked', 'Pclass'])['Fare'].mean().round(2)
avg_fare

```

mergeing, joining and concatenating

```python
courses = pd.read_csv('/home/development/Downloads/practice/courses.csv')
students = pd.read_csv('//home/development/Downloads/practice/students.csv')
nov = pd.read_csv('/home/development/Downloads/practice/reg-month1.csv')
dec = pd.read_csv('/home/development/Downloads/practice/reg-month2.csv')
courses.head()
```

```python
students.tail()
```

```python
# pd.concat
pd.concat([nov, dec])
```

```python
# ignore index
regs= pd.concat([nov, dec], ignore_index=True)
regs
```

```python
# Multiindex DataFrame
multi= pd.concat([nov, dec], keys=['Nov', 'Dec'])
multi
# multi.loc[('Dec', 4)]
```

Merge

```python
# inner join
pd.merge(students, regs, on='student_id', how='inner')
```

```python
# left join
pd.merge(courses, regs, on='course_id', how='left')
```

```python
# right join
employees = pd.DataFrame({
    'employee_id': [101, 102, 103],
    'name': ['Alice', 'Bob', 'Charlie']
})

projects = pd.DataFrame({
    'employee_id': [101, 102, 104], 
    'project_name': ['AI Research', 'Web Development', 'Cyber Security']
})

pd.merge(employees, projects, on='employee_id', how='right')
```

```python
# outer joiin 
pd.merge(employees, projects, on='employee_id', how='outer')
```

```python
# 1. find total revenue generated
total = pd.merge(courses, regs, on='course_id', how='inner')['price'].sum().item()
total
```

```python
# 2. find month by month revenue
revenue = pd.concat([nov, dec], keys=['Nov', 'Dec']).reset_index()
revenue.merge(courses, on='course_id', how='inner').groupby('level_0')['price'].sum()
```

```python
# 3. Print the registration table
# cols -> name -> course -> price
regs.merge(students, on='student_id').merge(courses, on='course_id')[['name', 'course_name', 'price']]
```

```python
# 4. Plot bar chart for revenue/course
regs.merge(courses, on='course_id').groupby('course_name').sum().plot(kind='bar')
```

```python
# 5. find students who enrolled in both the month
common_id = np.intersect1d(nov['student_id'], dec['student_id'])
common_id
students[students['student_id'].isin(common_id)]
```

```python
# 6. find course that got no enrollment
course_id = np.setdiff1d(courses['course_id'], regs['course_id'])
course_id
courses[courses['course_id'].isin(course_id)]
```

```python
# 7. find students who did not enroll into any courses
student_id_list = np.setdiff1d(students['student_id'], regs['student_id'])
student_id_list
students[students['student_id'].isin(student_id_list)]
```
